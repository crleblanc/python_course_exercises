# GeoNet FITS demo using Pandas
This demo uses the Pandas module to download and parse the CSV data from the FITS API.

```python
#!/usr/bin/env python
import pandas as pd
from urllib import parse

# use interactive notebook plots (zooming, panning, etc).
# Jupyter/Ipython specific, standalone scripts don't require this.
%matplotlib notebook

# use urlencode to encode URL, could also do this manually eg: params='networkID=%s&siteID=%s&typeID=%s' % ('LI', 'GISB', 'e')
params = parse.urlencode({'networkID':'LI', 'siteID':'GISB', 'typeID':'e'})
url = "https://fits.geonet.org.nz/observation?%s" % params
data = pd.read_csv(url, parse_dates=[0], index_col=0)
data.plot()
```

## read again with more appropriate arguments
* correctly parses the date and uses the date as an index
* resample to monthly 

```python
data = pd.read_csv(url, parse_dates=[0], index_col=0)

#rename columns
data.columns = pd.Index(['e', 'error'])

# resample to month and plot
resampled = data.resample('M').mean()

resampled.plot()
```
